# Use K-means to cluster the segments based on their location

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('sfpark_filtered_segments.csv',sep=';')
```

```python
data['x'] = (data['startx'] + data['endx'])/2
data['y'] = (data['starty'] + data['endy'])/2
segment_loc = np.array(data[['x','y']])
```

```python
from sklearn.cluster import KMeans
```

```python
centers_init = np.array([[-122.44, 37.80],
                        [-122.433, 37.785],
                        [-122.417, 37.807],
                        [-122.42, 37.778],
                        [-122.42, 37.76],
                        [-122.42, 37.796],
                        [-122.393, 37.793],
                        [-122.394, 37.78],
                        [-122.402, 37.79]])
kmeans = KMeans(n_clusters=9, init=(centers_init)).fit(segment_loc)
centers = kmeans.cluster_centers_
segment_pred = kmeans.predict(segment_loc)
# pd.DataFrame(centers).to_csv('centers.csv')
# pd.DataFrame(segment_pred).to_csv('prediction.csv')
```

```python
segment_loc
```

```python
segment_id = pd.DataFrame(data['segmentid'])
segments_0 = segment_id[segment_pred==0]
segments_0['x'] = centers[0,0]
segments_0['y'] = centers[0,1]
segments_1 = segment_id[segment_pred==1]
segments_1['x'] = centers[1,0]
segments_1['y'] = centers[1,1]
segments_2 = segment_id[segment_pred==2]
segments_2['x'] = centers[2,0]
segments_2['y'] = centers[2,1]
segments_3 = segment_id[segment_pred==3]
segments_3['x'] = centers[3,0]
segments_3['y'] = centers[3,1]
segments_4 = segment_id[segment_pred==4]
segments_4['x'] = centers[4,0]
segments_4['y'] = centers[4,1]
segments_5 = segment_id[segment_pred==5]
segments_5['x'] = centers[5,0]
segments_5['y'] = centers[5,1]
segments_6 = segment_id[segment_pred==6]
segments_6['x'] = centers[6,0]
segments_6['y'] = centers[6,1]
segments_7 = segment_id[segment_pred==7]
segments_7['x'] = centers[7,0]
segments_7['y'] = centers[7,1]
segments_8 = segment_id[segment_pred==8]
segments_8['x'] = centers[8,0]
segments_8['y'] = centers[8,1]
```

```python
segments_8
```

```python
plt.figure(figsize=(9,7))
plt.gca().set_facecolor('white')
plt.scatter(segment_loc[np.argwhere(segment_pred==0),0], segment_loc[np.argwhere(segment_pred==0),1], cmap='hsv', label='0')
plt.scatter(segment_loc[np.argwhere(segment_pred==1),0], segment_loc[np.argwhere(segment_pred==1),1], cmap='hsv', label='1')
plt.scatter(segment_loc[np.argwhere(segment_pred==2),0], segment_loc[np.argwhere(segment_pred==2),1], cmap='hsv', label='2')
plt.scatter(segment_loc[np.argwhere(segment_pred==3),0], segment_loc[np.argwhere(segment_pred==3),1], cmap='hsv', label='3')
plt.scatter(segment_loc[np.argwhere(segment_pred==4),0], segment_loc[np.argwhere(segment_pred==4),1], cmap='hsv', label='4')
plt.scatter(segment_loc[np.argwhere(segment_pred==5),0], segment_loc[np.argwhere(segment_pred==5),1], cmap='hsv', label='5')
plt.scatter(segment_loc[np.argwhere(segment_pred==6),0], segment_loc[np.argwhere(segment_pred==6),1], cmap='hsv', label='6')
plt.scatter(segment_loc[np.argwhere(segment_pred==7),0], segment_loc[np.argwhere(segment_pred==7),1], cmap='hsv', label='7')
plt.scatter(segment_loc[np.argwhere(segment_pred==8),0], segment_loc[np.argwhere(segment_pred==8),1], cmap='hsv', label='8')
plt.legend()
area = np.array([[200, 200]]*4)
plt.scatter(centers[:,0], centers[:,1], c='orange', marker='o', s=area)
```

```python
segments_0 = segments_0.rename(columns={'0':'segmentid'})
segments_1 = segments_1.rename(columns={'0':'segmentid'})
segments_2 = segments_2.rename(columns={'0':'segmentid'})
segments_3 = segments_3.rename(columns={'0':'segmentid'})
segments_4 = segments_4.rename(columns={'0':'segmentid'})
segments_5 = segments_5.rename(columns={'0':'segmentid'})
segments_6 = segments_6.rename(columns={'0':'segmentid'})
segments_7 = segments_7.rename(columns={'0':'segmentid'})
segments_8 = segments_8.rename(columns={'0':'segmentid'})
```

```python
segments_0.to_csv('cluster_area0.csv', index=False)
segments_1.to_csv('cluster_area1.csv', index=False)
segments_2.to_csv('cluster_area2.csv', index=False)
segments_3.to_csv('cluster_area3.csv', index=False)
segments_4.to_csv('cluster_area4.csv', index=False)
segments_5.to_csv('cluster_area5.csv', index=False)
segments_6.to_csv('cluster_area6.csv', index=False)
segments_7.to_csv('cluster_area7.csv', index=False)
segments_8.to_csv('cluster_area8.csv', index=False)
```
