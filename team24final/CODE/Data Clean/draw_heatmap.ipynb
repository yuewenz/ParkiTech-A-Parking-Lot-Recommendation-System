```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
geo_data = pd.read_csv('sfpark_filtered_segments.csv',sep=';')
park_data = pd.read_csv('sfpark_filtered_136_247_100taxis.csv',sep=';')
```

```python
plt.figure(figsize=(9,7))
data = []
for i in range(len(geo_data)):
    data.append(park_data[park_data['segmentid']==geo_data.iloc[i]['segmentid']].iloc[13]['occupied']/park_data[park_data['segmentid']==geo_data.iloc[i]['segmentid']].iloc[13]['capacity'])
plt.scatter((geo_data['startx'] + geo_data['endx']) / 2, (geo_data['starty'] + geo_data['endy']) / 2, cmap='RdYlBu', c = data)
```
